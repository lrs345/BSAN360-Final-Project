```python
#Bin Data:
#Asked the following prompt to ChatGPT - 'How to include 0 in a bin using pandas?'
#Since I am did not actually utilize any data from the dataset, there is no data
#protection needed while using AI in this case.

#After doing the bins, the use % columns become less relevant. I will not drop them
#for now in case I want to look into different cutoffs in the future.

#Also some pitchers do not utilize a certain type of pitch. This is not going to 
#be excluded from the data and will be instead label as 'No Use'. Their performance
#can still be measured in testing and the lack of use is valuable to determining the
#importance of the pitch type to success.

import numpy as np
bins = [-1,0,25,50,np.inf]
names = ['No Use','Minimal Use','Moderate Use','Primary Pitch']
pitch['Use of Fastball'] = pd.cut(pitch['Fastball Use %'], bins, labels=names)
pitch['Use of Breaking'] = pd.cut(pitch['Breaking Use %'], bins, labels=names)
pitch['Use of Offspeed'] = pd.cut(pitch['Offspeed Use %'], bins, labels=names)
pitch
```

```python
#Research Question 1 requires ANOVA.
#Research Questions 2 and 3 require regression.
#I assigned my categorical columns for pitch usage numerical values in order to run
#tests. The results below seemed a lot more efficient than getting dummy variables
#for each column.

#Asked ChatGPT: "how to assign a category to a numerical value in python"
#and clarified with "they are currently categories, I want to assign 
#them numerical values".

pitch['FB Use'] = pitch['Use of Fastball'].map({
    'No Use': 0,
    'Minimal Use': 1,
    'Moderate Use': 2,
    'Primary Pitch': 3
 })

pitch['Breaking Use'] = pitch['Use of Breaking'].map({
    'No Use': 0,
    'Minimal Use': 1,
    'Moderate Use': 2,
    'Primary Pitch': 3
 })

pitch['Offspeed Use'] = pitch['Use of Offspeed'].map({
    'No Use': 0,
    'Minimal Use': 1,
    'Moderate Use': 2,
    'Primary Pitch': 3
 })

pitchuse = pitch[[ 'FB Use', 'Breaking Use', 'Offspeed Use']]
pitchuse
```

```python
#Research Question 1 - running ANOVA Tests to compare ERA among pitch usages.

#Asked Copilot "I want to run an anova test to test the difference in one column in
#three different dataframes. How should I approach this?".

g1 = pitch[pitch['Use of Fastball'] == 'Primary Pitch']
g2 = pitch[pitch['Use of Breaking'] == 'Primary Pitch']
g3 = pitch[pitch['Use of Offspeed'] == 'Primary Pitch']

g1 = g1['ERA']
g2 = g2['ERA']
g3 = g3['ERA']

from scipy.stats import f_oneway

F, p = f_oneway(g1, g2, g3)

print("F-statistic:", F)
print("p-value:", p)

#not statistical significance between which pitch is a primary pitch and ERA.
```

```python
#Research Question 1 continued:
h1 = pitch[pitch['Use of Fastball'] == 'Moderate Use']
h2 = pitch[pitch['Use of Breaking'] == 'Moderate Use']
h3 = pitch[pitch['Use of Offspeed'] == 'Moderate Use']

h1 = h1['ERA']
h2 = h2['ERA']
h3 = h3['ERA']

F, p = f_oneway(h1, h2, h3)

print("F-statistic:", F)
print("p-value:", p)

#not statistical significance between moderate pitch usages and ERA.
```

```python
#Research Question 1 continued:
i1 = pitch[pitch['Use of Fastball'] == 'Minimal Use']
i2 = pitch[pitch['Use of Breaking'] == 'Minimal Use']
i3 = pitch[pitch['Use of Offspeed'] == 'Minimal Use']

i1 = i1['ERA']
i2 = i2['ERA']
i3 = i3['ERA']

F, p = f_oneway(i1, i2, i3)

print("F-statistic:", F)
print("p-value:", p)

#not statistical significance between minimal pitch usages and ERA.
```

```python
#Research Question 1 continued:
#No 'Use of Fastball' equals 'No Use'
j2 = pitch[pitch['Use of Breaking'] == 'No Use']
j3 = pitch[pitch['Use of Offspeed'] == 'No Use']

j2 = j2['ERA']
j3 = j3['ERA']

F, p = f_oneway(j2, j3)

print("F-statistic:", F)
print("p-value:", p)

print(j2.mean())
print(j3.mean())

#Looks like we finally found something somewhat statistically significant here 
#(p-value = 0.05). There is a difference in ERAs between pitchers who do
#not use Breaking pitches and pitchers who do not use offspeed pitches. This signals
#either that some use of breaking pitches leads to lower ERAs or less use of 
#offspeed may lead to lower ERAs. Let's test further.

#Next, I will prepare tests to compare among each fb group against each other, each
#breaking group against each other, and each offspeed group against each other.
#FB
F, p = f_oneway(g1, h1, i1)

print("F-statistic:", F)
print("p-value:", p)

#Breaking
F, p = f_oneway(g2, h2, i2, j2)

print("F-statistic:", F)
print("p-value:", p)

F, p = f_oneway(g3, h3, i3, j3)

#Offspeed
print("F-statistic:", F)
print("p-value:", p)

print(g3.mean())
print(h3.mean())
print(i3.mean())
print(j3.mean())

#It appears that there is statistcal significance to the difference in ERAs among
#the different groups of offspeed usage. No offspeed usage provides the best ERA
#and using offspeed as a primary pitch leads to the highest ERA. Therefore, there
#may be an advantage to limiting offspeed use.
```
